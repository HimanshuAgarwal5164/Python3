```python
import PIL
from PIL import Image
import inspect
from IPython.display import display
```

```python
img = Image.open(r"C:\Users\Himanshu Agarwal\Pictures\Saved Pictures\Me.jpg")
display(img)
```

```python
img = img.convert("RGB")
```

```python
from PIL import ImageEnhance
```

```python
enhancer = ImageEnhance.Brightness(img)
img1 = enhancer.enhance(0.2)
display(img1)
```

```python
enhancer = ImageEnhance.Brightness(img)
for i in range (0,11):
    img1 = enhancer.enhance(i/10)
    (w,h) = img.size
    img1 = img1.resize((int(w//10),int(h/10)))
    display(img1)
```
