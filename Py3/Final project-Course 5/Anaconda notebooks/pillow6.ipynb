```python
import PIL
from PIL import Image
from IPython.display import display
from PIL import ImageEnhance
```

```python
img = Image.open(r"C:\Users\Himanshu Agarwal\Pictures\Saved Pictures\Me.jpg")
img = img.convert("RGB")
(w,h) = img.size
enhancer = ImageEnhance.Brightness(img)
images=[]

for i in range (1,10):
    img1 = enhancer.enhance(i/10)
    images.append(img1)
contact_sheet = PIL.Image.new(img.mode,(w*3,h*3))
x=0
y=0
for a in images:
    contact_sheet.paste(a,(x,y))
    x = x+w
    if x == w*3:
        x = 0
        y = y+h

contact_sheet = contact_sheet.resize((w,h))
display(contact_sheet)
```
