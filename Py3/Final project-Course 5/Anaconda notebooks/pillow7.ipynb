```python
import PIL
from PIL import Image
from IPython.display import display
from PIL import ImageEnhance
```

```python
img = Image.open(r"C:\Users\Himanshu Agarwal\Pictures\Saved Pictures\Me.jpg")
(w,h) = img.size
r,g,b = img.split()
images = []
k = 0.1
for i in range (0,9):
    if i < 3:
        r1 = r.point(lambda x:x*k)
        result = Image.merge("RGB",(r1,g,b))
        images.append(result)
        k = k+0.4
        if k >= 1:
            k = 0.1
    elif i < 6:
        g1 = g.point(lambda x:x*k)
        result = Image.merge("RGB",(r,g1,b))
        images.append(result)
        k = k+0.4
        if k >= 1:
            k = 0.1
    else:
        b1 = b.point(lambda x:x*k)
        result = Image.merge("RGB",(r,g,b1))
        images.append(result)
        k = k+0.4
        if k >= 1:
            k = 0.1
contact_sheet = PIL.Image.new(img.mode,(w*3,h*3))
x = 0
y = 0
for a in images:
    contact_sheet.paste(a,(x,y))
    x = x+w
    if x == w*3:
        x = 0
        y = y+h

contact_sheet = contact_sheet.resize((w,h))
display(contact_sheet)    
    
```
