```python
import PIL
from PIL import Image
from IPython.display import display
from PIL import ImageEnhance
```

```python
img = Image.open(r"C:\Users\Himanshu Agarwal\Pictures\Saved Pictures\Me.jpg")
(width,height) = img.size
enhancer = ImageEnhance.Brightness(img)
images = []
for i in range (0,11):
    img1 = enhancer.enhance(i/10)
    images.append(img1)
contact_sheet = PIL.Image.new(img.mode,(width,height*11))
x=0
y=0
for i in images:
    contact_sheet.paste(i,(x,y))
    y=y+height
contact_sheet = contact_sheet.resize((width//5,height*2))
display(contact_sheet)
```
